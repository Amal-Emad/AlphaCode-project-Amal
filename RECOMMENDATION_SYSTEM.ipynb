```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/dataset.csv')

# Display the first few rows of the dataset
print(data.head())

# Handle missing values (if any)
data.fillna(0, inplace=True)

# Convert duration from milliseconds to minutes
data['duration_minutes'] = data['duration_ms'] / 60000
```

```python
# Shape of the dataset (number of rows and columns)
print("Shape of the dataset:", data.shape)

# Summary statistics of numerical columns
print("Summary statistics of numerical columns:")
print(data.describe())

# Information about the dataset
print("Information about the dataset:")
print(data.info())

# Unique values in categorical columns
print("Unique values in categorical columns:")
for column in data.select_dtypes(include=['object']):
    print(column, ":", data[column].nunique())

# Distribution of genre labels
plt.figure(figsize=(10, 6))
sns.countplot(data['track_genre'])
plt.title('Distribution of Genre Labels')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap of numerical features
# Drop non-numeric columns before generating the correlation heatmap
numeric_data = data.select_dtypes(include=[np.number])

# Correlation heatmap of numerical features
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()

```

```python
# Convert all values in the 'track_genre' column to strings
data['track_genre'] = data['track_genre'].astype(str)

# Ensure there are no missing values in the 'track_genre' column
data['track_genre'].fillna('', inplace=True)

# Retry encoding the 'track_genre' column
label_encoder = LabelEncoder()
data['track_genre_encoded'] = label_encoder.fit_transform(data['track_genre'])



# Select relevant features
features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
            'duration_minutes', 'popularity']

# Normalize numerical features
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

# Encode categorical features
label_encoder = LabelEncoder()
data['track_genre_encoded'] = label_encoder.fit_transform(data['track_genre'])

# Split the data into training and testing sets
X = data[features]
y = data['track_genre_encoded']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
classifiers = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Support Vector Machine": SVC(kernel='linear', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "k-Nearest Neighbors": KNeighborsClassifier(n_neighbors=5)
}

# Train and evaluate classifiers
performance_metrics = {}

for name, clf in classifiers.items():
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)



    performance_metrics[name] = {
        "accuracy": accuracy,

    }

    print(f"Classifier: {name}")
    print(f"Accuracy: {accuracy}")

    print("-" * 80)
```

```python
# Extract accuracy scores for plotting
accuracy_scores = [metrics["accuracy"] for metrics in performance_metrics.values()]
classifiers_names = list(classifiers.keys())

# Plotting the accuracy scores
plt.figure(figsize=(10, 6))
sns.barplot(x=classifiers_names, y=accuracy_scores)
plt.title('Classifier Accuracy Comparison')
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.ylim(0, 0.5)
plt.show()
```

```python
# @title Default title text
from sklearn.metrics.pairwise import cosine_similarity
# Function to recommend tracks
def recommend_tracks(input_track_id, data, model, features, top_n=10):
    # Get the input track features
    input_track = data[data['track_id'] == input_track_id]
    if input_track.empty:
        print("Track ID not found in the dataset.")
        return pd.DataFrame()

    input_features = input_track[features].values

    # Predict the genre of the input track
    predicted_genre = model.predict(input_features)[0]

    # Get all tracks with the predicted genre
    similar_tracks = data[data['track_genre_encoded'] == predicted_genre]

    # Calculate similarity between input track and similar tracks
    similar_tracks['similarity'] = cosine_similarity(input_features, similar_tracks[features].values).flatten()

    # Sort tracks by similarity and return top N recommendations
    recommendations = similar_tracks.sort_values(by='similarity', ascending=False).head(top_n)

    return recommendations[['track_id', 'track_name', 'artists', 'album_name', 'similarity']]

input_track_id = '61T3ukOW74kzMkVy5pOLDR'  # Replace with an actual track ID from your dataset

recommendations_dict = {}

for name, clf in classifiers.items():
    recommendations = recommend_tracks(input_track_id, data, clf, features, top_n=5)
    recommendations_dict[name] = recommendations

# Display recommendations in a formatted HTML layout
html_content = """
<style>
    body {
        font-family: Arial, sans-serif;
        background-color: #f4f4f4;
        color: #333;
        padding: 20px;
    }
    .classifier-section {
        background-color: #fff;
        border-radius: 10px;
        padding: 20px;
        margin-bottom: 20px;
        box-shadow: 0 0 10px rgba(0,0,0,0.1);
    }
    .classifier-title {
        font-size: 24px;
        color: #2c3e50;
        margin-bottom: 10px;
    }
    .recommendations {
        list-style-type: none;
        padding: 0;
    }
    .recommendation {
        background-color: #ecf0f1;
        border-radius: 5px;
        padding: 10px;
        margin-bottom: 10px;
    }
    .track-name {
        font-size: 18px;
        font-weight: bold;
    }
    .artists, .album-name, .similarity {
        font-size: 14px;
        color: #7f8c8d;
    }
</style>
"""

for name, recommendations in recommendations_dict.items():
    html_content += f"""
    <div class="classifier-section">
        <div class="classifier-title">{name} Recommendations</div>
        <ul class="recommendations">
    """
    for _, row in recommendations.iterrows():
        html_content += f"""
        <li class="recommendation">
            <div class="track-name">{row['track_name']}</div>
            <div class="artists">Artists: {row['artists']}</div>
            <div class="album-name">Album: {row['album_name']}</div>
            <div class="similarity">Similarity: {row['similarity']:.2f}</div>
        </li>
        """
    html_content += "</ul></div>"

# Display the styled HTML
from IPython.display import display, HTML
display(HTML(html_content))
```
